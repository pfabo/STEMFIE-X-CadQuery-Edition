## <font color='purple'> <b> Basic library classes </b></font>

**Pulley**

    Pulley(type, holes=True, fill=False, thick=0.8, beams=3)

**Pulley Holder**

    Pulley_Holder_1(height)
    Pulley_Holder_2(height)

**Simple Wheels**

    Wheel(r, h=1/4, d=1/2 )
        r - diameter in BU, range [1,2]
        h - height in BU
        d - shaft offset

## <font color='Purple'> <b> Usage - Pulleys and Holders</i> </b> </font>

```python
from lib import *

N = 2
p1 = Pulley(N)
p1.export_step('./step/pull_02_01')

p2 = Pulley_Holder_1(N)
p2.export_step('./step/comp_PH_01_02')

# assembly pulley and holders
p1 = p1.Rx().BU_T([1/2, -1/2, N])
p3 = p2.copy().Rz(180).BU_Txy([1,-1])
p1.U([p2,p3]).Rx(-90).Ry(30)

display(SVG(p1.obj.toSvg()))
```

```python
N = 3
p1 = Pulley(N)
p1.export_step('./step/pull_03_01')

p2 = Pulley_Holder_1(N)
p2.export_step('./step/comp_PH_01_03')

p1.Rx().BU_T([1/2, -1/2, N])
p3 = p2.copy().Rz(180).BU_Txy([1,-1])
p1.U([p2,p3]).Rx(-90).Ry(30)

display(SVG(p1.obj.toSvg()))
```
