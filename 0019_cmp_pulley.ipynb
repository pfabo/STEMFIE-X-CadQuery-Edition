### <font color='teal'> Examples </i></b> </font>

```python
from lib import *

N = 2
p1 = Pulley(N).Rx().BU_T([1/2, -1/2, N])
p2 = Pulley_Holder_1(N)
p3 = p2.copy().Rz(180).BU_Txy([1,-1])
p1.U([p2,p3]).Rx(-90).Ry(30)

convert_to_image(p1,'./img/pulley_01')
Image('./img/pulley_01.png', width=800)
```
