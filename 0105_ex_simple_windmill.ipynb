### <font color='purple'> <b> Component generation program </b></font>

Program for generating STL components from the STEMFIE-X library. 



```Python
import sys
sys.path.append("./../../notebook")

from lib import *
from numpy import pi, sin, cos, arcsin
import cadquery as cq
from cadquery import exporters

w1 = Wheel(1)
w1.obj.export('whell_1.stl')
w1.obj.export('whell_1.step')

b5 = Brace(5)
b5.obj.export('brace_5.stl')
b5.obj.export('brace_5.step')

b10 = Brace(10)
b10.obj.export('brace_10.stl')
b10.obj.export('brace_10.step')

bh = Beam_H_Block(10, 5, 1/4, 1/4)
bh.obj.export('bh_block.stl')
bh.obj.export('bh_block.step')

# chord of a part of a circle 5 BU 
phi = arcsin(2.5/3)/pi*180
dx = 3*cos(phi/180*pi)
ba = Brace_Arc(3, phi*2, 1/4, 5).Rz(180-phi).BU_Txy(dx, 0)
ba.obj.export('ba_brace.stl')
ba.obj.export('ba_brace.step')

p2 = Pulley(2)
p2.obj.export('p2_pulley.stl')
p2.obj.export('p2_pulley.step')

p15 = Pulley(1.5)
p15.obj.export('p15_pulley.stl')
p15.obj.export('p15_pulley.step')
```
