```python
from lib import *
from numpy import arcsin, pi

D = 6   # D >= 2*R+1  
R = 3   # arc radius

alpha = arcsin( (D-1) / (2*R ) )
beta = 2*alpha
H = R*cos(alpha)
bdeg = beta/pi*180     # convert rad -> deg
h = R*cos(alpha)
br1 = Brace_Arc(R, bdeg, 1/4, 5, center=True);  br1.export_step('./model//step_02_small_windmill/brace_A_03_14_06_112')  # 2 pcs
display(SVG(br1.obj.toSvg()))
```
