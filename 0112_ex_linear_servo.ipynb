### <font color='teal'> <b> Components </b></font>

The critical element is the gear on the servo axis. The standard diameter of the axis is 4.8mm, but it can vary between servo manufacturers. At the same time, the parameters of the wheel depend on the material used and the 3D printing, speed and temperature. Therefore, it is advisable to print several gears with a slightly modified diameter and put the wheel on the shaft without much force.

STEP files of the generated linear servo  components are located in the directory **./step/servo** and/or **./stl/servo**.


```python
from lib import *

class Servo_9g_Gear(Stemfie_X):
    def __init__(self, ds):   
        '''
        ds - servo axis 
        '''
        Stemfie_X.__init__(self)  
        sg = Spur_Gear(20, bore=2)

        d1 = 0.5
        h1 = 0.25
        c1 = BU_Cylinder( d1, h1, hole=False).BU_Tz(h1/2)

        d2 = 0.24
        h2 = 0.2
        c2 = BU_Cylinder(d2, h2, hole=False).BU_Tz(h1 + h2/2)

        d3 = ds                             # servo axe
        h3 = 0.5
        c3 = BU_Cylinder(d3, h3, hole=False).BU_Tz(h1 + h2 + h3/2)

        d4 = 1
        h4 = 0.25
        c4 = BU_Cylinder(d4, h4, hole=False).BU_Tz(1/2+h4/2)

        c1.U([c2, c3])
        sg.U(c4)
        sg.D(c1)
        self.obj = sg.obj

sg = Servo_9g_Gear(0.47);  sg.export_step('./step/servo/spur_gear_470');  sg.export_stl('./stl/servo/spur_gear_470'); 
sg = Servo_9g_Gear(0.48);  sg.export_step('./step/servo/spur_gear_480');  sg.export_stl('./stl/servo/spur_gear_480');
sg = Servo_9g_Gear(0.49);  sg.export_step('./step/servo/spur_gear_490');  sg.export_stl('./stl/servo/spur_gear_490');

sg.Rz().Rx(15)
display(SVG(sg.obj.toSvg()))
```

```python
# linear gearing
rr = Rack_Gear(6, width = 1, height=1/4).Rx().Rz().BU_Ty(-3).BU_Tz(1/2).BU_Tx(-1/2)
bb = Brace(9, 1/4, holes=False).Rz().BU_Ty(-4)
hh = Hole_List([ [0, 0], [0, 8] ], length = 2).BU_Ty(-4)
bb.D(hh)
rr.U(bb)
rr.export_step('./step/servo/rack_gear');  
rr.export_stl('./stl/servo/rack_gear'); 

rr.Rz(-80)
display(SVG(rr.obj.toSvg()))
```
