Set of basic elementary components in BU coordinates. Generally usable as add-on components and a basis for the construction of derived components.

## <font color='purple'> <b> Library classes </b></font>

    BU_Cube([x,y,z], center)
    BU_Cylinder(diameter, radius, angle, hole, center)
    BU_Polyline([ [x1,y1], [x2,y2], ...], height)
    BU_Bar(length, center)
    BU_Sphere(diameter, center)                        [ToDo]
    BU_Cone(diam1, diam2, height, angle, center)       [ToDo]    

### <font color='teal'> <b> Component labeling </b></font>

For better orientation in directories, it is advisable to label the generated components (STL, STEP files) with a name that describes their parameters. Unused and default parameters are not written for clarity.

    base_t_p1_p2_p3_ ...                p1 ... pn  the number of parameters depends on the component type

    t  - component type                 pn - parameters, in BU units or angle in deg (ppp)
         B - cube                            01 =    1 BU            or number of points
         C - cylinder                           ...
         P - polyline                        10 =   10 BU
         R - bar                             12 =  1/2 BU 
         Q - sphere                          14 =  1/4 BU
         N - cone
         X - user defined  

## <font color='purple'> <b>Usage </b> </font>

```python
from lib import *

bc = BU_Cube([3,2,1], False)     # base_B_03_02_01      3x2x1 BU cube

Image(convert_to_image(bc,'./img/base_B_03_02_01'), width=500)
```

```python
b1 = BU_Cylinder(1,1, hole=True, center=False)                # base_C_01_01
b2 = BU_Cylinder(1,2, hole=False, center=False, angle=270)    # base_C_01_02_270

b1.U(b2.BU_Tx(2) )
Image(convert_to_image(b1,'./img/bu_demo_cylinder'), width=500)
```

```python
cc = [ [0,0], [1,0], [2,1], [1,2], [0,2], [1,1], [0,0] ]
b1 = BU_Polyline(cc, 1)      # base_P_07_01

Image(convert_to_image(b1,'./img/base_P_07_01'), width=500)
```
