### <font color='teal'> <i><b>Examples</i></b> </font>

```python
from IPython.display import Image
from lib import *

# BU_Cube()

b1 = BU_Cube(1, False)           # 1x1x1 BU cube
b2 = BU_Cube([3,2,1], False)     # 3x2x1 BU cube

b2.BU_Tx(2)                      # translate and unite for display in Jupyter
b1.U(b2)
convert_to_image(b1,'./img/bu_cube')
Image('./img/bu_cube.png', width=500)
```

```python
# BU_Cylinder()

b1 = BU_Cylinder(1,1, hole=True, center=False)
b2 = BU_Cylinder(1,2, hole=False, center=False, angle=270)


b2.BU_Tx(2) 
b1.U(b2)#.Rx()

convert_to_image(b1,'./img/bu_cylinder')
Image('./img/bu_cylinder.png', width=500)
```

```python
# BU_Polyline()

cc = [ [0,0], [1,0], [2,1], [1,2], [0,2], [1,1], [0,0] ]
b1 = BU_Polyline(cc, 1)

convert_to_image(b1,'./img/bu_poly')
Image('./img/bu_poly.png', width=500)
```
