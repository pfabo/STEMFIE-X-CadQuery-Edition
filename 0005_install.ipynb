[STEMFIE](https://www.stemfie.org/) is a construction kit for children, authored by [Paulo Kiefe](https://www.stemfie.org/about). The idea of the project is based on the design and 3D printing of universal components and connecting elements and the subsequent construction of own models.

For the parametric design of kit components (beams, braces ...) of various shapes and sizes [Cantareus](https://github.com/Cantareus/Stemfie_OpenSCAD/blob/main/docs/stemfie.scad.md) created a parametric library for openSCAD program. Other graphic programs and design CAD systems such as FreeCAD, Blender, and Tinkercad can also be used to design and construct models.

The **STEMFIE-X** kit uses the basic principles of the **STEMFIE** project with modifications for use as a platform for creating more complex structures and models using mechanical and electronic elements - motors, servos, electronic circuits and control by microcontrollers. The building kit is primarily intended for polytechnic teaching at secondary schools and the first semesters of universities in the teaching of parametric modeling, CAD systems, constructions of machines and mechanisms. In STEMFIE-X the base unit (BU) is adjusted to a size of 10mm and the hole diameters are changed to 4mm, standard M4 screws are used as connecting components. 

**STEMFIE-X CadQuery Edition** is implementation of the STEMFIE-X library for Python on the CadQuery platform (the original STEMFIE-X library is implemented in *OpenSCAD* and is based on a functional approach). The **STEMFIE-X CadQuery Edition** library consists of a set of classes, with the help of which it is possible to create and modify basic as well as derived components of the kit. When creating models, it is possible to use the connection of elementary components or by combining them, it is possible to create specialized components using the Python program. The designed components can be generated as STEP or STL files and printed on a 3D printer using standard procedures.

## <font color='purple'> <b> STEMFIE-X Usage</b></font>

STEMFIE-X can be used in different ways

* for easy generation of standard components from library classes (**Beam_Block, Brace ...**) and creating STL or STEP files for 3D printing
* for editing and modifying standard components and creating custom components using logical operations **union, difference, intersection- U(),D(),I()**
* for creating structural units using transformation operations e.g. translation and rotation operations **BU_Tx, Rx() ...**

Generated STEP or STL files can be further used in other design programs e.g. FreeCAD.

## <font color='purple'> <b> STEMFIE-X Documentation</b></font>

Documentation for using the library is in the format Jupyter Notebook. The library files are located in the **./lib** directory in this notebook. The library can be used directly from a Jupyter Notebook environment or as a standard library in any Python programming environment.

**Using STEMFIE-X Library in Jupyter Notebook**

```python
# library import
from lib import *          

# creating of components, rotation and translation
b1 = Beam_Block([3, 2 ,1/2], [False, False, True]).Rx(-45)
b2 = Brace(5).BU_Tx(4)
b1.U([b1,b2])

# creating and displaying an image
Image(convert_to_image(b1,'./img/demo'), width=400)
```
