## <font color='purple'> <b> Library classes </b></font>

    Wheel(r, h=1/4, d=1/2 )
        r - diameter in BU, range [1,2,3]
        h - height in BU
        d - shaft offset


### <font color='teal'> <b> Component labeling </b></font>

    wheel_nn_hh_dd_bb       basic form, unlisted parameters have default values

    nn - wheel radius in BU units
         01, 02, 03
         
    hh - wheel thickness in BU units, default 1/4 BU
         14 - 1/4 BU
         12 - 1/2 BU
         01 -   1 BU
         02 -   2 BU
         ...

    dd - wheel hub length default 1/2 BU
         14 - 1/4 BU
         12 - 1/2 BU
         01 -   1 BU
         02 -   2 BU
         ...

    bb - number of arm, only for radius 03
        03 - 6 arms
        04 - 8 arms
