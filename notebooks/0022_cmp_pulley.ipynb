## <font color='purple'> <b> Basic library classes </b></font>

**Pulley**

    Pulley(type, holes=True, fill=False, thick=0.8, beams=3)

**Pulley Holder**

    Pulley_Holder_1(height)
    Pulley_Holder_2(height)

**Simple Wheels**

    Wheel(r, h=1/4, d=1/2 )
        r - diameter in BU, range [1,2]
        h - height in BU
        d - shaft offset

```python
from lib import *

N = 2
p2 = Pulley(N)
p2.export_step('./step/pulley/pull_02_01')

ph2 = Pulley_Holder_1(N)
ph2.export_step('./step/pulley/comp_PH_01_02')

p2 = p2.Rx().BU_T([1/2, -1/2, N])
ph3 = ph2.copy().Rz(180).BU_Txy([1,-1])
p2.U([ph2,ph3]).Rx(-90).Ry(30)

display(SVG(p2.obj.toSvg()))
```
