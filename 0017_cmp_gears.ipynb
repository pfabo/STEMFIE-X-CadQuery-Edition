# <font color='navy'> <i><b> Gears</i></b> </font>

The Gears library uses the [CQ_Gears](https://github.com/meadiode/cq_gears) generator.

### <font color='teal'> Examples </i></b> </font>

```python
from lib import *
from cq_gears import SpurGear

class Spur_Gear(Stemfie_X):
    
    def __init__(self, teeth_number=20, width=1/2, bore=2):
        module=1.0
        spur_gear = SpurGear(module, teeth_number, width=self.BU*width, bore_d=bore)
        self.obj = cq.Workplane("XY" , origin=(0,0,0))
        
        self.obj = (self.obj).gear(spur_gear)


g1 = Spur_Gear(10, 1/2)
g2 = Spur_Gear(30, 1/2).Rz(360/30/2).BU_Tx(2)
g3 = Spur_Gear(10, 1/2).BU_Tx(2).BU_Tz(1/2)
g4 = Spur_Gear(30, 1/2).Rz(360/30/2).BU_Tx(4).BU_Tz(1/2)

g1.U([g2,g3,g4])

convert_to_image(g1,'./img/gears_01')
Image('./img/gears_01.png', width=600)
```
