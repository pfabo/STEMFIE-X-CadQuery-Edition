The *./lib/gears.py* library is used to create gears and their accessories according to DIN780. The original OpenSCAD library is on [github-e](https://github.com/chrisspen/gears/blob/master/gears.scad), transcription to CadQuery is on [CQ_Gears](https://github.com/meadiode/cq_gears).

The basic type is a spur gear. For spur gears (*spur gear*) with $modul=1$, the axial distance in BU units for two gears with the number of teeth $n_1$ and $n_2$ is given by the relation

\begin{equation*}
d = \frac{n_1 + n_2}{20}
\end{equation*}

## <font color='purple'> <b> Library classes </b></font>

    Spur_Gear(teeth_number=20, width=1/2, bore=HR)
    Rack_Gear(length, width=1/2, height=1/2)
    

### <font color='teal'> Usage </i></b> </font>

```python
from lib import *

sg = Spur_Gear(20)
display(SVG(sg.obj.toSvg()))
```

```python
sg1 = Spur_Gear(15)
sg2 = Spur_Gear(25)    # (25+15)/20 = 2BU
bbb = Brace(3).BU_Tz(1/2)
sg2.BU_Tx(2)
sg1.U([sg2, bbb])
display(SVG(sg1.obj.toSvg()))
```

```python
sg3 = Spur_Gear(30)
sg4 = Spur_Gear(10).BU_Tz(1/2)
sg3.U(sg4)

sg5 = Spur_Gear(30).BU_Tz(1/2).BU_Tx(2)
bbb = Brace(3).BU_Tz(1)
sg3.U([sg5, bbb])
display(SVG(sg3.obj.toSvg()))
```

```python
rg1 = Rack_Gear(5, 1)
sg7 = Spur_Gear(20).BU_Ty(1).BU_Tx(2).BU_Tz(1/4)
bbb = Brace(7,1/4).Rx().BU_T([-1/2, -1/4-1/8, 1/2])
rg1.U([bbb, sg7])
display(SVG(rg1.obj.toSvg()))
```
