The basic Beam_Block components is used to create beams and plates. Modifications of the H and U blocks can be used as bases for small models and functional units.

<img src="./img/image_block.png" width="800">

## <font color='purple'> <b> Basic library classes </b></font>

    Beam_Block([x,y,z], center)
    Beam_Block(d, center)
    Beam_U_Block(x,y)
    Beam_H_Block(x,y)
    Beam_L_Block(x,y)

### <font color='teal'> <b> Component labeling </b></font>

    block_t_xx_yy_zz_hhh    basic form
    block_t_xx_yy           abbreviated forms
    block_t_xx

    t  - block type                               xx,yy,zz - block dimension
         B - simple beam                                     01 =   1 BU
         U - beam block modifications                        ...
         H                                                   10 =  10 BU
         L                                                   12 = 1/2 BU
         X - user defined non standard block                 14 = 1/4 BU 
        
    hhh - holes configuration in x y z directions
        000 - no holes
        100 - holes in x direction
        ...
        111 - holes in all directions

    block_B_xx_yy_zz_hhh
    block_H_xx_yy_hx_hy
    block_U_xx_yy_hx_hy
    block_L_xx_yy_hx_hy

    Examples
        
    
        block_B_05          -> block_B_05_01_01_111    
                               standard block Beam_Block([5,1,1], holes=[True,True,True])
        block_B_05_03       -> block_B_05_03_14_001    
                               block plate Beam_Block([5,3,1/4], holes=[False,False,True])


## <font color='purple'> <b> Usage </b></font>

```python
from lib import *

b1 = Beam_Block(5)                # General Beam_Block
display(SVG(b1.obj.toSvg()))
```

```python
b2 = Beam_Block([3, 2 ,1], [True, True, True]) # 2x3 Beam_Block
b2.BU_Tx(4).BU_Tx(4)
display(SVG(b2.obj.toSvg()))
```

```python
b3 = Beam_Block([4, 3 ,1/2], [False, False, True])  # 4x3 BU plate created from Beam_Block
display(SVG(b3.obj.toSvg()))
```

```python
b4 = Beam_U_Block(4,3).Rx(45)   # 4x3 BU U-Shape plate 
Image(convert_to_image(b4,'./img/block_04'), width=600)
```

```python
b5 = Beam_H_Block(4,3).Rx(45)   # 4x3 BU U-Shape plate 
display(SVG(b5.obj.toSvg()))
```
