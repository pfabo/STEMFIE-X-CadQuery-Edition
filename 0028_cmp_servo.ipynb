The analog servo is available under various designations (SG90, 9g, GO-09, MG90S ...) and in various designs with plastic or metal gears. The servo axis has a diameter of 4.8mm, but it may vary between manufacturers. The servo is controlled by a pulse width modulated signal (PWM).

<img src="./img/msg90s_cable.png" width="200">
<img src="./img/mg90s_dim.jpg" width="600">

### <font color='teal'> <b> Library classes </b></font>

    Servo_9g()           servo model         ./step/servo/servo_9g.step
                                             ./stl/servo/servo_9g.stl'
    
    Servo_9g_Holder()    servo holder        ./step/servo/servo_9g_holder.step
                                             ./stl/servo/servo_9g_holder.stl

The servo model is oriented so that the axis origin is at position (0,0) BU.

<img src="./img/msg90s_model.png" width="300">


### <font color='purple'> <b> Usage </b></font>

```python
from lib import *
s = Servo_9g()  
s.export_step('./step/servo/servo_9g')
s.export_stl('./stl/servo/servo_9g')

s.Rz().Rx(15)
Image(convert_to_image(s,'./img/demo_servo'), width=600)
```

```python
sh = Servo_9g_Holder()
sh.export_step('./step/servo/servo_9g_holder')
sh.export_stl('./stl/servo/servo_9g_holder')

sh.Rz().Rx(15)
Image(convert_to_image(sh,'./img/demo_servo_holder'), width=400)
```

```python
s2 = Servo_9g()
s2.BU_T([1/8, 1/2, 1])      # servo position in the holder
h2 = Servo_9g_Holder()
h2.U(s2).Rx(-90).Ry(-80)
Image(convert_to_image(h2,'./img/demo_servo_in_holder'), width=500)
```
