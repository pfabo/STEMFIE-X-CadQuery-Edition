### <font color='teal'> <b> Example -  Linear servo </b></font>

STEP files of the generated linear servo  components are located in the directory ./step/servo

```python
class Servo_9g_Gear(Stemfie_X):
    def __init__(self, ds):   
        '''
        ds - servo axis 
        '''
        Stemfie_X.__init__(self)  
        sg = Spur_Gear(20, bore=2)

        d1 = 0.5
        h1 = 0.25
        c1 = BU_Cylinder( d1, h1, hole=False).BU_Tz(h1/2)

        d2 = 0.24
        h2 = 0.2
        c2 = BU_Cylinder(d2, h2, hole=False).BU_Tz(h1 + h2/2)

        d3 = ds                             # servo axe
        h3 = 0.5
        c3 = BU_Cylinder(d3, h3, hole=False).BU_Tz(h1 + h2 + h3/2)

        d4 = 1
        h4 = 0.25
        c4 = BU_Cylinder(d4, h4, hole=False).BU_Tz(1/2+h4/2)

        c1.U([c2, c3])
        sg.U(c4)
        sg.D(c1)
        self.obj = sg.obj

sg = Servo_9g_Gear(0.49)
sg.Rz().Rx(15)
Image(convert_to_image(sg,'./img/servo_gear'), width=400)
```
