The Stemfie_X class methods represent, in shorthand, basic operations for manipulating objects. 
Using these operations, you can modify basic components or create composite and derived ones.

### <font color='teal'> <b> Translations </b></font>

    BU_Tx(n)                      # translate object in x,y,z direction in BU units
    BU_Ty(n)
    BU_Tz(n)
    BU_Txy([x,y])  BU_Txy(x,y)
    BU_Tyz([x,y])  BU_Tyz(y,z)
    BU_Tzx([z,x])  BU_Tzx(z,x)
    BU_T([x,y,z])  BU_T(x,y,z)
    Tx(d)                         # translate object in [mm] units
    Ty(d)
    Tz(d)
    T([x,y,z])     T(x,y,z)
  
### <font color='teal'> <b> Rotations </b></font>

    Rx(angle)
    Ry(angle)
    Rz(angle)

### <font color='teal'> <b> Mirroring and Mirroring with copy </b></font> 

    Mx()
    My()
    Mz()
    MKx()
    MKy()
    MKz()  

### <font color='teal'> <b> Boolean operations </b></font> 

    U(c)   U([c1,c2 ...])     union
    D(c)   D([c1,c2 ...])     difference
    I(c)   I([c1,c2 ...])     intersection     

## <font color='Purple'> <b>Usage </i></b> </font>

```python
from lib import *                         # U-Shape Beam Block
b1 = Beam_Block([3,1,1])                  # components, rotating and shifting
b2 = Beam_Block([3,1,1]).Ry()
b3 = Beam_Block([2,1,1]).Ry().BU_Tx(2)

b1.U([b2,b3])                             # b1 = b1 + b2 + b3
b1.Rx(30)                                 # rotate for view

Image(convert_to_image(b1,'./img/op1'), width=600)
```
