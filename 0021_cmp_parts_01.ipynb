## <font color='teal'> <b> Movable Pulley </b></font>

````python
from lib import *

# Free pulley on braces - for crane arms

p1 = Pulley(1, holes=True, fill=False, thick=0.45 ).Rx()

b1 = Brace(5).Rx().BU_Ty(-1/4)
b2 = b1.copy().Mx()
b3 = Brace(2, 1/2).Rx().BU_Ty(1/4).BU_Tx(2)

# main beam
b4 = Brace(7).Rx().Ry().BU_Tz(3).BU_Ty(-1/2).BU_Tx(4)
b5 = b4.copy().Mx()

# axes
br1 = BU_Bar(2).Rx()
br2 = BU_Bar(2).Rx().BU_Tx(4)

b1.U([p1, b1, b2, b3, b4, b5, br1, br2]).Rx().Ry(135)


convert_to_image(b1,'./img/parts_01')
Image('./img/parts_01.png', width=800)
````
