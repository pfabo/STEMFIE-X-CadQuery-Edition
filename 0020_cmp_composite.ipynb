# <font color='navy'> <i><b> Composite components</i></b> </font>

Using Python programming, you can create composite components for your designs using the Stemfie-X library.

### <font color='teal'> Example </i></b> </font>

```python
from lib import *

# Construction of Brace with slot

bs = Brace(5, holes=False)
hh = Hole_Slot(3)
hs = Hole_List([ [3,0], [4,0]] )
bs.D([hh, hs]) 

convert_to_image(bs,'./img/composite')
Image('./img/composite.png', width=300)
```

```python
from lib import *

# Triangle Brace

n = 5
b1 = Brace(n)
b2 = Brace(n).Rz(60)
b3 = Brace(n).Rz(120).BU_Tx(n-1)

b1.U([b2, b3]) 

convert_to_image(b1,'./img/composite_02')
Image('./img/composite_02.png', width=500)
```
