Using Python programming, you can create composite components for your designs using the Stemfie-X library. Simple composite components can be created using the basic operations of union **U()**, difference **D()** and intersection **I()**.

### <font color='teal'> Examples </i></b> </font>

```python
from lib import *

# Brace with slot and holes

bs = Brace(5, holes=False)          # brace without holes
hh = Hole_Slot(3)                   # slot
hs = Hole_List([ [3,0], [4,0]] )    # holes
bs.D([hh, hs]).Rz(30)                      # difference 

display(SVG(bs.obj.toSvg()))
```

```python
# Symetric brace with slot

bs = Brace(6, holes=False)          # brace without holes
hh = Hole_Slot(4).BU_Tx(1)          # slot + shift
hs = Hole_List([ [0,0], [5,0]] )    # holes
bs.D([hh, hs]).Rz(30)                       # difference 

display(SVG(bs.obj.toSvg()))
```

```python
# triangle Brace

n = 5   # size of brace in BU units
b1 = Brace(n)
b2 = Brace(n).Rz(60)
b3 = Brace(n).Rz(120).BU_Tx(n-1)

b1.U([b2, b3]) 

display(SVG(b1.obj.toSvg()))
```

```python
# bent brace

b1 = Brace(5, 1/2)
b2 = Brace(3, 1/2).Rz(120)

b1.U(b2) 

display(SVG(b1.obj.toSvg()))
```

```python
b1 = Beam_Block([5, 5, 1/2], [False, False,True], center=True)
b2 = Beam_Block([3, 3, 1/2], [False, False, False], center=True)

b1.D(b2) 

display(SVG(b1.obj.toSvg()))
```
