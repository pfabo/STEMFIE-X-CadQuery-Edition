# <font color='navy'> <i><b> Composite components</i></b> </font>

Using Python programming, you can create composite components for your designs using the Stemfie-X library. Simple composite components can be created using the basic operations of union U(), difference D() and intersection I().

### <font color='teal'> Examples </i></b> </font>

```python
from lib import *

# brace with slot and holes

bs = Brace(5, holes=False)          # brace without holes
hh = Hole_Slot(3)                   # slot
hs = Hole_List([ [3,0], [4,0]] )    # holes
bs.D([hh, hs])                      # difference 

convert_to_image(bs,'./img/composite_01')
Image('./img/composite_01.png', width=300)
```

```python
# symetric brace with slot

bs = Brace(6, holes=False)          # brace without holes
hh = Hole_Slot(4).BU_Tx(1)          # slot + shift
hs = Hole_List([ [0,0], [5,0]] )    # holes
bs.D([hh, hs])                      # difference 

convert_to_image(bs,'./img/composite_02')
Image('./img/composite_02.png', width=300)
```

```python
# triangle brace

n = 5   # size of brace in BU units
b1 = Brace(n)
b2 = Brace(n).Rz(60)
b3 = Brace(n).Rz(120).BU_Tx(n-1)

b1.U([b2, b3]) 

convert_to_image(b1,'./img/composite_03')
Image('./img/composite_03.png', width=500)
```

```python
# bent brace

b1 = Brace(5, 1/2)
b2 = Brace(3, 1/2).Rz(120)

b1.U(b2) 

convert_to_image(b1,'./img/composite_04')
Image('./img/composite_04.png', width=500)
```

```python
# bent brace

b1 = Brace(5, 1)
b2 = Brace(3, 1/2).Rz().BU_Tz(1/4)

b1.U(b2).Ry(20) 

convert_to_image(b1,'./img/composite_05')
Image('./img/composite_05.png', width=600)
```
