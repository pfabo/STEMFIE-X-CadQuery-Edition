### <font color='teal'> <i><b>Example</i></b> </font>

```python
from IPython.display import Image
from lib import *

b1 = BU_Cube([1,1,1], False).BU_Ty(-2).BU_Tx(-2)
b2 = BU_Cylinder(1,1, hole=True, center=False).BU_Tx(-2)
b3 = BU_Cylinder(1,1, hole=False, center=False, angle=180).BU_Tx(-2).BU_Ty(1+1/2)

pp = [ [0,0], [1,0], [2,1], [1,2], [0,2], [1,1], [0,0] ]
b4 = BU_Polyline(pp, 1).BU_Ty(-3)

b1 = b1.U([b2,b3,b4]).Rx().Ry()

convert_to_image(b1,'./img/basic_comp')
Image('./img/basic_comp.png', width=500)
```
